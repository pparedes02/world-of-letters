```python
# https://instaloader.github.io/as-module.html#python-module-instaloader
import instaloader

# Get instance
L = instaloader.Instaloader(download_videos=False, 
                            download_video_thumbnails=False)
```

```python
from instaloader import Profile
from datetime import datetime
from itertools import dropwhile, takewhile
```

```python
USERNAME1 = ['scooterscoffee', 'dunkin', 'pukupukucafe','dunnbroscoffee',
             'estacion329cafe', 'biggbycoffee', 'coffeebeanery1976', 
             'gloriajeanscoffees', 'thehumanbeancoffee' ]
             
for i in USERNAME1:
    profile = Profile.from_username(L.context, i)
    posts = profile.get_posts()
    SINCE = datetime(2019, 10, 5)
    UNTIL = datetime(2017, 1, 1)
    for post in takewhile(lambda p: p.date > UNTIL, 
                          dropwhile(lambda p: p.date > SINCE, posts)):
        L.download_post(post, target=profile.username)
```
